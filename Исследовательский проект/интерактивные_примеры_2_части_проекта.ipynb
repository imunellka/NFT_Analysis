```python
import plotly
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots
import numpy as np
import pandas as pd
nft_dataset = pd.read_excel('games_final_corrected_2_new.xlsx', index_col='Index')
nft_dataset
```

```python

fig = go.Figure(data=[go.Histogram(x=nft_dataset['BasePrice'])])
fig.update_layout(legend_orientation="h",
                  legend=dict(x=.5, xanchor="center"),
                  title="Зависимость количества игр от цены",
                  xaxis_title="Цена",
                  yaxis_title="Количество игр",
                  margin=dict(l=0, r=0, t=30, b=0))
fig.show()
```

##Вывод:
Количество игр с низкой стоимостью значительно преобладает над премиум сегментом. 

```python
sum_counts = nft_dataset.groupby('minRAM(GB)').size()
fig = go.Figure()
fig.add_trace(go.Pie(values=sum_counts, labels=sum_counts.index))
fig.update_layout(title="Круговая диаграмма по минимальному объему ОЗУ")
fig.show()
```

##Вывод: 
4.0 и 8.0 ГБ ОЗУ по популярности преобладают перед другими размерами ОЗУ
